# Linear Chain of Atoms

## Here, we investigate a model for a linear chain of atoms with free ends.

For a 1D chain of atoms, of different masses, we can use two approximations:
- The interactions are quadratic due to expanding the potential around its minimum (equilibrium position) without pursuing higher orders (harmonic oscillator potential)
- and they only exist between the nearest neighbours

In writing the Newtonian equations of motion for such a system we obtain

$$
m_i \delta\ddot{x_i} = - ( k_{i-1} + k_i ) \delta x_i + k_{i-1} \delta x_{i-1} + k_i \delta x_{i+1}
$$

which are the equations of motion for the displacement of each atom from its equilibrium position.

We can express these equations using matrices:

$$
\mathbf{M} \, \delta\ddot{\mathbf{x}} = - \mathbf{K} \, \delta\mathbf{x}
$$

where $\mathbf{M}$ is positive definite diagonal matrix, which means it's invertible, so, we can alternatively express the equation as follows

$$
\delta\ddot{\mathbf{x}} = - \mathbf{M}^{-1} \mathbf{K} \, \delta\mathbf{x} \tag{1}
$$

for that all $k_i$'s are all positive, the matrix $\mathbf{K}$ is garaunteed to be positive semi-definite, thus, the matrix $\mathbf{M}^{-1} \mathbf{K}$ is garaunteed to possess a complete set of eigenvectors in searching for ones with eigenvalues of the form $\lambda=-\omega^2$.

Since the vector $\delta\ddot{\mathbf{x}}$ is just a vector that belong to the space spanned by the eigenvectors of the matrix $\mathbf{M}^{-1} \mathbf{K}$, it suffices to search for its eigenvectors and any general solution to $Eq. (1)$ can be expressed as a superposition of those eigenvectors, or the eigenmodes of the system in this case, so to speak.

So, we only need to solve the equation

$$
det\left( \mathbf{M}^{-1} \mathbf{K} - \omega^2 \mathbf{I} \right) = 0
$$

which is what the following code does using numpy!

```python
import numpy as np
import matplotlib
matplotlib.rcParams['legend.fancybox'] = True
import matplotlib.pyplot as plt
import scipy.integrate
import scipy.optimize
import scipy.constants
```

```python
class Chain :             # for a finite chain of atoms of number N
    def __init__( self, N, m, k ):
        self.N = N        # the number of atoms
        self.m = np.array(m)        # the ordered list of masses
        self.k = np.array(k)        # the ordered list of spring constants

        # initializing the eigenfrequencies and eigenmodes
        self.eigfreq = np.zeros((self.N))
        self.eigmod = np.zeros((self.N, self.N))
        
    def __call__( self ) :
        # constructing the mass matrix and its inverse
        M = np.diag(self.m)         # the mass-matrix
        invM = np.diag(1/self.m)    # the inverted mass matrix

        # constructing the spring-constant matrix
        K = np.zeros( (self.N,self.N) )

        # Diagonal elements
        diag = np.zeros(self.N)
        diag[:-1] += self.k        # contributes from left spring
        diag[1:] += self.k         # contributes from right spring
        np.fill_diagonal(K, diag)
    
        # Off-diagonal elements
        K[np.arange(self.N - 1), np.arange(1, self.N)] = -1 * self.k  # upper diagonal
        K[np.arange(1, self.N), np.arange(self.N - 1)] = -1 * self.k  # lower diagonal

        invM_K = invM @ K

        # Calculating the eigenvalues and eigenvectors
        eigenvalues, eigenvectors = np.linalg.eigh(invM_K)    # (eigh) is used to ensure the return of real eigenvalues (h --> hermitian, which invM_K indeed is)
        eigenvalues = np.clip(eigenvalues, 0, None)           # to ignore small numbers (especially if negative)

        # the frequency values
        omegas = np.sqrt(eigenvalues)

        self.eigfreq = omegas
        self.eigmod = eigenvectors
        
        return omegas, eigenvectors

    def plot_frequencies(self):
        plt.figure(figsize=(6, 4))
        plt.plot(self.eigfreq, 'o-')
        plt.xlabel("Mode index")
        plt.ylabel("Frequency (rad/s)")
        plt.title("Normal Mode Frequencies")
        plt.grid(True)
        plt.show()

    def plot_normal_modes(self):
        plt.figure(figsize=(8, 2 * self.N))
        for i in range(self.N):
            plt.subplot(self.N, 1, i+1)
            plt.plot(self.eigmod[:, i], 'o-')
            plt.title(f'Mode {i+1}')
            plt.ylabel('Displacement')
            plt.grid(True)
        plt.xlabel('Particle index')
        plt.tight_layout()
        plt.show()
```

```python
c = scipy.constants.c
N_A = scipy.constants.N_A
```

```python
mC = ( 12e-3 / N_A )
mO = ( 16e-3 / N_A )
```

```python
kCO = 0.5 * mO * ( c ** 2 ) * ( ( 1353.6 * 1e-2 ) ** 2 )
kCC = 0.5 * mC * ( c ** 2 ) * ( ( 1855.1 * 1e-2 ) ** 2 )
```

# A. For Carbon suboxide (C3O2)

```python
N = 5
m = np.array([mO,mC,mC,mC,mO])
k = np.array([kCO,kCC,kCC,kCO])

chain = Chain(N, m, k)
omegas, eigenvectors = chain()

print("Frequencies:\n",omegas, "\nNormal Modes:\n", eigenvectors)

chain.plot_frequencies()
chain.plot_normal_modes()
```

# B. For C9O2

```python
N = 11
m = np.array([mO,mC,mC,mC,mC,mC,mC,mC,mC,mC,mO])
k = np.array([kCO,kCC,kCC,kCC,kCC,kCC,kCC,kCC,kCC,kCO])

chain = Chain(N, m, k)
omegas, eigenvectors = chain()

print("Frequencies:\n",omegas, "\nNormal Modes:\n", eigenvectors)

chain.plot_frequencies()
chain.plot_normal_modes()
```

The following sources were utilized to estimate the spring constants given the vibrational frequencies of the bonds scaled by the speed of light
- https://cccbdb.nist.gov/expvibs2x.asp
- http://www2.ess.ucla.edu/%7Eschauble/MoleculeHTML/CO2_html/CO2_page.html

The $k_i$'s were estimated using the formula

$$
k_i = \frac{1}{2}m_ic^2\kappa^2
$$

where $\kappa$ is the vibrational frequency scaled by $c$, as this equation is derivable from the case of two oscillating masses connected by one spring in between, which is a special case from the equations of motion overlaid at the beginning of this notbook.
