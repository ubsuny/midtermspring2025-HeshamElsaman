# ODEs: Cosmic inflation

Starting with the two equation:

$$
\begin{align}
H^2 &= \left( \frac{\dot{R}}{R} \right)^2 = \frac{8\pi G}{3} \rho \tag{1} \\
\dot{\rho} &= -3H\left(1 + w\right)\rho \tag{2}
\end{align}
$$

$$
\begin{align}
\Longrightarrow 2 H \dot{H} &= \frac{8 \pi G}{3} \dot{\rho} \\
&= \frac{8 \pi G}{3} \cdot -3H\left(1+w\right) \rho\\
&= -8 \pi G H \left(1+w\right) \cdot \frac{3}{8 \pi G} H^2\\
&= -3 H \left(1+w\right) H^2
\end{align}
$$

So, we end up with

$$
\frac{\dot{H}}{H^2} = \frac{-3}{2} \left(1+w\right) \tag{3}
$$

We can directly integrate $Eq. (3)$ using separation of variables as follows:

$$
\begin{align}
\frac{dH}{H^2} &= -\frac{3}{2} \left(1+w\right) dt \\
\int \frac{1}{H^2} \, dH &= \int -\frac{3}{2} \left(1+w\right) \, dt \\
\frac{-1}{H} &= -\frac{3}{2} \left(1+w\right) t - C
\end{align}
$$

So, we end up with

$$
H = \frac{1}{\frac{3}{2} \left(1+w\right) t + C} \tag{4}
$$

We basically arrived at the solution for $\rho$ which can be expressed using $Eqs. (1) \& (4)$ as:

$$
\rho (t) = \frac{3}{8 \pi G} \frac{1}{\left(\frac{3}{2} \left(1+w\right) t + C\right)^2} \tag{5}
$$

Now using $Eqs. (1) \& (4)$ again, we can solve for $R$ using the separation of variables again as follows:

$$
\begin{align}
\frac{\dot{R}}{R} &= \frac{1}{\frac{3}{2} \left(1+w\right) t + C} \\
\frac{dR}{R} &= \frac{dt}{\frac{3}{2} \left(1+w\right) t + C} \\
\int\frac{1}{R}\,dR &= \int\frac{1}{\frac{3}{2} \left(1+w\right) t + C}\,dt \\
\ln{\left(R\right)} &= \frac{1}{\frac{3}{2}\left(1+w\right)}\ln{\left(\frac{3}{2}\left(1+w\right)t+C\right)}+\ln{\left(B\right)}\\
&= \ln{\left(B\left[\frac{3}{2}\left(1+w\right)t+C\right]^{\frac{1}{\frac{3}{2}\left(1+w\right)}}\right)}
\end{align}
$$

So, we arrive at the solution for $R$ expressed as follows:

$$
R(t) = B\left[\frac{3}{2}\left(1+w\right)t+C\right]^{\frac{1}{\frac{3}{2}\left(1+w\right)}} \tag{6}
$$

Now, we're in a position to impose initial/boundary conditions.

For imposing initial conditions at some $t=t_0$, we impose that

$$
R(t_0) = 1 \Longrightarrow B = \left[\frac{3}{2}\left(1+w\right)t_0+C\right]^{\frac{-1}{\frac{3}{2}\left(1+w\right)}}
$$

thus,

$$
R(t) = \frac{\left[\frac{3}{2}\left(1+w\right)t+C\right]^{\frac{1}{\frac{3}{2}\left(1+w\right)}}}{\left[\frac{3}{2}\left(1+w\right)t_0+C\right]^{\frac{1}{\frac{3}{2}\left(1+w\right)}}}
$$

and from $Eq. (5)$ we get

$$
\rho_0 = \frac{3}{8 \pi G} \frac{1}{\left(\frac{3}{2} \left(1+w\right) t_0 + C\right)^2} \\
\Longrightarrow \boxed{C = \sqrt{\frac{3}{8 \pi G \rho_0}} - \frac{3}{2} \left(1+w\right) t_0}
$$

so, we obtain

$$
\rho(t) = \frac{3}{8 \pi G} \frac{1}{\left( \frac{3}{2}\left(1+w\right)t + \sqrt{\frac{3}{8 \pi G \rho_0}} - \frac{3}{2}\left(1+w\right)t_0 \right)^2} \tag{7}
$$

$$
R(t) = \left[ \frac{ \frac{3}{2}\left(1+w\right)t + \sqrt{ \frac{3}{8 \pi G \rho_0} } - \frac{3}{2}\left(1+w\right)t_0 }{ \frac{3}{2}\left(1+w\right)t_0 + \sqrt{ \frac{3}{8 \pi G \rho_0} } - \frac{3}{2}\left(1+w\right)t_0 } \right]^{ \frac{1}{\frac{3}{2}\left(1+w\right)} } \tag{8}
$$


Further, based on physical considerations, it's reasonable to assume the following boundary condition

$$
\text{as} \space t \longrightarrow 0 , \rho \longrightarrow \infty
$$

so, that condition imposes that $C=0$, and thus

$$
\rho_0 = \frac{3}{8 \pi G} \frac{1}{\left(\frac{3}{2} \left(1+w\right) t_0\right)^2}\\
$$

so, we end with the exact solutions

$$
\rho (t) = \frac{3}{8 \pi G} \frac{1}{\left( \frac{3}{2} \left(1+w\right) t \right)^2} \tag{9}
$$

$$
R(t) = \left( \frac{t}{t_0} \right)^{\frac{1}{\frac{3}{2}\left(1+w\right)}} \tag{10}
$$

Now, for scaling purposes, we set $t_0=1$ (representing the current moment), so we obtain our final set of equations

$$
\rho (t) = \frac{3}{8 \pi G} \frac{1}{\left( \frac{3}{2} \left(1+w\right) t \right)^2} \tag{11}
$$

$$
R(t) = \left( t \right)^{\frac{1}{\frac{3}{2}\left(1+w\right)}} \tag{12}
$$

```python
import numpy as np
import matplotlib
matplotlib.rcParams['legend.fancybox'] = True
import matplotlib.pyplot as plt
import scipy.integrate
import scipy.optimize
```

```python
params_0 = {
    't0':1.0,
    'dt':0.0001,
    'accuracy':1e-6,
    'G':1.0,
    'method':'RK45',
    'tmax':200.,
    'R0':1.0,
    'w':0.0,
}

params_1 = {
    't0':1.0,
    'dt':0.0001,
    'accuracy':1e-6,
    'G':1.0,
    'method':'RK45',
    'tmax':200.,
    'R0':1.0,
    'w':1.0,
}

params_2 = {
    't0':1.0,
    'dt':0.0001,
    'accuracy':1e-6,
    'G':1.0,
    'method':'RK45',
    'tmax':200.,
    'R0':1.0,
    'w':-1.0/3.0,
}
```

```python
class scale_factor :
    def __init__( self, params , RHO0=None ):
        self.params = params
        for iparam,param in params.items():
            setattr(self, iparam, params[iparam])
        if RHO0:
            self.rho0 = RHO0
        else:
            self.rho0 = ( 3 / ( 8 * self.G * np.pi ) ) * ( ( 2 / ( 3 * ( 1 + self.w ) ) ) ** 2 )
        self.c = np.sqrt( 3 / ( 8 * np.pi * self.G * self.rho0 ) ) - ( 3 / 2 ) * ( 1 + self.w ) * self.t0
    
    def set_param(self, parname, parval ):
        setattr( self, parname, parval )

    def eq_motion(self, t, u):
        Rdot = u[0] * np.sqrt( u[1] * ( ( 8 * np.pi * self.G ) / 3 ) )
        rhodot = -3 * ( Rdot / u[0] ) * ( 1 + self.w ) * u[1]
        return np.array([Rdot, rhodot])

    def Rt(self, t):
        # return t ** ( 2 / ( 3 * ( 1 + self.w ) ) ) # for c = 0
        return ( ( ( 3 / 2 ) * ( 1 + self.w ) * t + self.c ) / ( ( 3 / 2) * ( 1 + self.w ) * self.t0 + self.c ) ) ** ( 1 / ( ( 3 / 2 ) * ( 1 + self.w ) ) )

    def rhot(self, t):
        # return 1 / ( 6 * np.pi * self.G * ( ( ( 1 + self.w ) * t ) ** 2 ) ) # for c = 0
        return ( 3 / ( 8 * np.pi * self.G ) ) * ( ( 1 / ( ( 3 / 2 ) * ( 1 + self.w ) * t + self.c ) ) ** 2 )
    
    def analytical_solution(self, t):
        return np.array([Rt(t), rhot(t)])

    def __call__(self) :
        res = scipy.integrate.solve_ivp(self.eq_motion,t_span=(self.t0,self.tmax),
                                y0=np.array([self.R0, self.rho0]), 
                                vectorized=True,
                                atol=1e-8,rtol=1e-6,
                                method=self.method,
                                first_step = self.dt)
        return res
```

```python
s_factor_0 = scale_factor(params_0)

s_factor_0_res = s_factor_0()

t = s_factor_0_res.t
states = s_factor_0_res.y
R = states[0]
rho = states[1]

fig = plt.figure(1,figsize=(6,6))
plt.scatter(t,R, label="Numerical")
plt.plot(t,s_factor_0.Rt(t), label="Analytical", color='r')
plt.xlabel('time')
plt.ylabel('R(t)')
plt.legend()
plt.show()

fig = plt.figure(1,figsize=(6,6))
plt.scatter(t,rho, label="Numerical")
plt.plot(t,s_factor_0.rhot(t), label="Analytical", color='r')
plt.xlabel('time')
plt.ylabel('rho(t)')
plt.legend()
plt.show()
```
